```python
df['Percentage']=df['Total']/8
df['Percentage']
```

```python
def getgrade(Percentage):
    if Percentage>75:
        return 'A'
    elif (Percentage>=60 and Percentage<75):
        return 'B'
    else:
        return 'C'
df["Grade"]=df["Percentage"].apply(getgrade)
df
```
